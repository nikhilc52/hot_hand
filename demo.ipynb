```python
import pandas as pd
import duckdb
from tqdm import tqdm
```

```python
results = []
for i in tqdm(range(2000)):
    x = np.random.choice([0,1],size=20)
    y = np.concat([[np.nan],x[:-1]])
    z = np.concat([[np.nan]*2,x[:-2]])
    a = np.concat([[np.nan]*3,x[:-3]])
    df = pd.DataFrame({'x':x,'y':y,'z':z,'a':a})
    df = duckdb.sql("""SELECT * FROM df
                        WHERE y = 1""").df()
    if pd.isna(df['x'].mean()): continue
    results.append(df['x'].mean())
```

```python
2*(np.mean(results)-.5)
```
